**Question 1**: Design a sentiment analysis application using OpenAI's models to determine the
sentiment of a given text (positive, negative, or neutral)?

```python
!pip install openai
```

```python
import openai
openai.api_key=""

def senti_analysis(text):
  messages=[ {"role":"system", "content":"""You are to determine the sentiment of the given text. If you are not able to determine
                 the sentiment tell the the user that the statement is unclear. """},
      {"role":"user", "content":f""" You are to determine the sentimental analysis of the statement and give a single word
         answer either positive or negative:{text} """}
      ]
  response=openai.chat.completions.create(model="gpt-3.5-turbo",
                                           messages=messages,
                                           max_tokens=1,
                                           n=1,
                                           temperature=0)
  respond=response.choices[0].message.content.strip().lower()
  return  respond

prompt= input("Enter your sentence here....")
print("The sentence is ",senti_analysis(prompt))
```

Question 2: Design a Python program that compares two text inputs and determines their
similarity using the OpenAI API.
Instructions:
● Prompt the user to input two texts for comparison.
● Utilize the OpenAI GPT-3.5-turbo language model to analyze the similarity between the
provided texts.
● Display the similarity score to the user, indicating how similar the two texts are.
● Handle errors gracefully, such as empty inputs or API request failures.
● Ensure the program is properly documented and easy to understand.
● Optionally, provide additional information to the user about the comparison process or
● the meaning of the similarity score.

```python
import openai

openai.api_key=""

def check_similarity(text1,text2):
  messages=[ {"role":"system","content":"""You are to compare the two given texts and check for similarity between them.
                          Based on the similarity between them, provide a similarity score which shows the extent of similarity.
                          Display the similarity score to the user, indicating how similar the two texts are."""},
             {"role":"user", "content":f"""You are to check for similarity between the given texts and give a similarity score,
                                        in numerical value, between them:{text1,text2}"""}
  ]
  response=openai.chat.completions.create(model="gpt-3.5-turbo",
                                           messages=messages,
                                           max_tokens=1,
                                           n=1,
                                           temperature=0 )
  respond=response.choices[0].message.content.strip().lower()
  return respond

prompt1=input("Input text here...")
prompt2=input("Input text here..")
response=check_similarity(prompt1,prompt2)
print("The value of similarity between them is ",response)
```
