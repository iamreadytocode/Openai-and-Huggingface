```python
from transformers import AutoModelForCausalLM, AutoTokenizer
model_id="Unbabel/TowerBase-7B-v0.1"
token= AutoTokenizer.from_pretrained(model_id)

# print(token.decode(outputs[o],skip_special_tokens=True))
```

```python
model = AutoModelForCausalLM.from_pretrained(model_id)

text= "English:My name is Abdullah.\nJapanese:"
inputs=token(text,return_tensors="pt")
outputs=model.generate(**inputs,max_new_tokens=20)
```

```python
print(token.decode(outputs[0],skip_special_tokens=True))
```
