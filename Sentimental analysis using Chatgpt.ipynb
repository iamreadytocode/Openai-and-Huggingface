```python
import openai
openai.api_key=''
def sentimental(text):
  messages = [
   {"role":"system","content":"""you are trained to analyse and detect yhe sentiments of my statement.
    if you can't find out simply say can't tell and check manually """},
    {"role":"user","content":f"""you are supposed to tell if the statement is positive or negative.
    just give your output in one word by either printing positive or negative:{text} """}
    ]
  response=openai.ChatCompletion.create(
                  model="gpt-3.5-turbo",
                  messages=messages,
                  max_tokens=1,
                  n=1,
                  stop=None ,
                  temperature=0)

  response_text= response.choices[0].message.content.strip().lower()

  return response_text


```

```python
inp="I love ALLAH"
print(inp,": The sentiment is ",sentimental(inp))
```
